```python
print("Finding location where max repetition occurs:")
print("")
loc=[]
for i in range(len(Freq_Seq)):
    if Freq_Seq[i]==2:
        loc.append(i)
print(loc)
```

```python
def reverse(s):
    return s[::-1]
```

```python
def allindex(seq,item):
    start_at = -1
    locs = []
    while True:
        try:
            loc = seq.index(item,start_at+1)
        except ValueError:
            break
        else:
            locs.append(loc)
            start_at = loc
    return locs
```
